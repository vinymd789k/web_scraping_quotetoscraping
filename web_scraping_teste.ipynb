```python
# Bibliotecas.
import pandas as pd
import requests 
from bs4 import BeautifulSoup
```

```python
# Url do site.
url = 'http://quotes.toscrape.com/'

# Requisição.
resposta = requests.get(url)
# HTTP.
print(resposta.status_code)
```

```python
# Imprimindo conteúdo.
print(resposta.headers)
```

```python
print(resposta.content)
```

```python
# Arnazenando conteúdo.
html = resposta.content

# HTML para beautifulsuop.
html_soup = BeautifulSoup(html, 'html.parser')
html_soup
```

```python
# Procurar tag das noticias.
cartao = html_soup.find_all('div', attrs={'class': 'quote'})
cartao
```

```python
# Lista com as tags das frases.
list_frase = []

# Itinerar em cada cartao para retirar a tag com título
for frase in cartao:
    phrase = frase.find('span', {'class': 'text'})
    list_frase.append(phrase)

list_frase
```

```python
# Lista somente com textos.
list_frase_texto = []

# Retirar somente os textos.
for texto in list_frase:
    phrase_text = (texto.text)
    list_frase_texto.append(phrase_text)

list_frase_texto
```

```python
# Procurando class author
list_author = []

for author in cartao:
    author_encontrado = author.find('small', {'class': 'author'})
    list_author.append(author_encontrado)

list_author
```

```python
# Apenas o texto dos autores
list_author_texto = []

for texto in list_author:
    author_texto = (texto.text)
    list_author_texto.append(author_texto)

list_author_texto
```

```python
# Link dos autores.
list_link_author = []

for link in cartao:
    link_author = link.find('small', {'class': 'author'}).find_next_sibling('a')['href']
    list_link_author.append(link_author)

list_link_author
```

```python
# Dicionário.
dict_web = {'frase': list_frase_texto, 'autor': list_author_texto, 'link': list_link_author}
dict_web
```

```python
# Dataframe.
df_web = pd.DataFrame(dict_web)
df_web
```

```python
# Convertendo CSV
df_web.to_csv('banco_de_dados_frases.csv')
```
